```python
def j_insertion_between_i_and_vowel(word): # This produces [aijon]* so I need to before this turn aion to ajon so this does not happen here
    matches = re.finditer(r'[^a]i[\-]?[\*]?[eaøɛoʌu]+', word) #if /ai/ then insertion would create an extra j, so I exclude a preceding /a/
    padding = 0 # when a j is inserted, the length is increased by one. As the length increases, the padding increases
    for match in matches:
        #print(match)
        #print(match.span()[1]) #span turns the match into a tuple, out of that we want the last index
        changed_index = match.span()[1]  - 1 + padding # minus one gives the vowel before which a j will be inserted
        #print(changed_index)
        word = word[:changed_index] + 'j' +word[changed_index:]
        #print(word)
        padding += 1
    return(word)
#j_insertion_between_i_and_vowel('i-e-ion')   
print(j_insertion_between_i_and_vowel('ia'))
```

```python
def me_change(word): 
    if match:=re.search('-me-[\*]?[p|n]', word):
       word = word.replace('-me-','-m-')  
    elif match:=re.search('-me-[\*]?[rb|nd]', word): #ʌrbindan and ʌndʌstan
        word = re.sub('-me-','-', word)   
    elif match:=re.search('-me-', word):
        word = word.replace('-me-','-n-')
    return word
print(me_change('u-me-pir'))
print(me_change('a-me-xir'))
print(me_change('o-me-nix'))
print(me_change('a-me-rbk'))
all_verbs['Generated Verb-SR'] = all_verbs['Generated Verb-SR'].apply(me_change)
all_verbs.head(5)
```

```python
def e_raising_in_den(word): # this rule occures before prog e rounding, so it is written above that rule
    pattern1 = 'den\-i'
    pattern2 = 'den\-\*i'
    if match:=re.search(pattern1, word):
        word = re.sub(pattern1,'din-i', word) 
    if match:=re.search(pattern2, word):
        word = re.sub(pattern2,'din-*i', word)   
    return word

print(e_raising_in_den('u-zden-i'))
print(e_raising_in_den('u-zden-*i'))
all_verbs['Generated Verb-SR'] = all_verbs['Generated Verb-SR'].apply(e_raising_in_den)
all_verbs.to_csv('./all_verbs_den_din.csv', encoding = 'utf-8-sig', index = False)
all_verbs = pd.read_csv('all_verbs_den_din.csv')
all_verbs.head()
```

```python
def d_deletion_in_den(word):
    pattern = '[d|t][\*]?\-d[e|i]n\-'
    if match:=re.search(pattern, word):
        index = match.span()[1]
        word = word[:index-4] + '' + word[index-3:] 
    return word 
print(d_deletion_in_den('d-den-e'))
print(d_deletion_in_den('t*-den-e'))
print(d_deletion_in_den('t-den-e')) 
all_verbs['Generated Verb-SR'] = all_verbs['Generated Verb-SR'].apply(d_deletion_in_den)
all_verbs.to_csv('./all_verbs_d_deletion.csv', encoding = 'utf-8-sig', index = False)
all_verbs = pd.read_csv('all_verbs_d_deletion.csv')
all_verbs.head()
```

```python
def vowel_deletion(word): 
    pattern = 'e[\*]?\-[\*]?[ieɛaøuoʌ]+' 
    if match:=re.search(pattern, word):
        index = match.span()[0]
        word = word[:index] + '' + word[index+1:] 
    return word
print(vowel_deletion('*ma*-ʌvid͡ʒ-∅'))




all_verbs['Generated Verb-SR'] = all_verbs['Generated Verb-SR'].apply(vowel_deletion)
all_verbs.to_csv('./all_verbs_vowel_deletion.csv', encoding = 'utf-8-sig', index = False)
all_verbs = pd.read_csv('all_verbs_vowel_deletion.csv')
all_verbs.head()
```

```python
def e_raising_in_prefixes(word): # i better put /e/ deletion above this
    # rather than letting this form be and deleting first i: because there are verbs in which i-i are the case and i di not want to delete the first i. 
    # an example is me-ger-i-i 'you are blocked!'
    pattern1 = '\*ne\*\-[^ieɛaøuoʌ:]+i' # i included - due to me-*ne*-t:-i so the change does not apply with two -s
    pattern2 = '\*(?:be|de|me)+\*\-[\*]?[^ieɛaøuoʌ:]+i'
    pattern3 = '(?:be|de|me)+\-[^ieɛaøuoʌ:]+i'
    if match:=re.search(pattern1, word):
        index = match.span()[0] #.span() return a tuple containing the starting and ending index of the matched string
        word = word[:index+2] + 'i' + word[index+3:]
    if match:=re.search(pattern2, word):
        index = match.span()[0]
        word = word[:index+2] + 'i' + word[index+3:]
    elif match:=re.search(pattern3, word):
        index = match.span()[0]
        word = word[:index+1] + 'i' + word[index+2:]
    return word

print(e_raising_in_prefixes('*ne*-di'))
print(e_raising_in_prefixes('me-di'))
print(e_raising_in_prefixes('me-*ni*-di'))
print(e_raising_in_prefixes('me-*ne*-nd͡ʒ-i'))
print(e_raising_in_prefixes('*be*-lbin-i-em(ɛ)'))
print(e_raising_in_prefixes('me-lbin-i-em(ɛ)'))
print(e_raising_in_prefixes('ʌ-*ne*-jar-i-ast-abindɛ'))


all_verbs['Generated Verb-SR'] = all_verbs['Generated Verb-SR'].apply(e_raising_in_prefixes)
all_verbs.to_csv('./all_verbs_raised_prefix.csv', encoding = 'utf-8-sig', index = False)
all_verbs = pd.read_csv('all_verbs_raised_prefix.csv')
all_verbs.head()
```

```python
def e_backing(word):
    pattern = '(?:be|de|me|ne)[\*]?\-[^i͡eɛaøuoʌn]+o' 
    if match:=re.search(pattern, word):
        index = match.span()[0]
        word =word[:index+1]+ 'o' + word[index+2:]
    return word
print(e_backing('me-*ne*-Goʃ'))
print(e_backing('*ne*-dom-gh'))
print(e_backing('be-χ:o-*ast*-im(ɛ)'))
all_verbs['Generated Verb-SR'] = all_verbs['Generated Verb-SR'].apply(e_backing)
all_verbs.to_csv('./all_verbs_e_backing.csv', encoding = 'utf-8-sig', index = False)
all_verbs = pd.read_csv('all_verbs_e_backing.csv')
all_verbs.head()
```

```python
def regressive_e_rounding(word):
    pattern = 'e[^ieɛaøuoʌ]+ø' 
    while match:=re.search(pattern, word):
        index = match.span()[0]
        word = word[:index] + 'ø' + word[index+1:]
    return word
print(regressive_e_rounding('e-gø-gh'))
print(regressive_e_rounding('be-røt'))
print(regressive_e_rounding('bei-røt'))
all_verbs['Generated Verb-SR'] = all_verbs['Generated Verb-SR'].apply(regressive_e_rounding)
all_verbs.to_csv('./all_verbs_reg_rounding.csv', encoding = 'utf-8-sig', index = False)
all_verbs = pd.read_csv('all_verbs_reg_rounding.csv')
all_verbs.head()

```

```python
def progressive_e_rounding(word):
    pattern = 'ø[^ieɛaøuoʌ]+e'
    while match:=re.search(pattern, word):
        index = match.span()[-1]
        word = word[:index-1] + 'ø'+ word[index:]
    return word
print(progressive_e_rounding('øcdf-e'))
print(progressive_e_rounding('øddd-en-e'))
all_verbs['Generated Verb-SR'] = all_verbs['Generated Verb-SR'].apply(progressive_e_rounding)
all_verbs.to_csv('./all_verbs_prog_rounding.csv', encoding = 'utf-8-sig', index = False)
all_verbs = pd.read_csv('all_verbs_prog_rounding.csv')
all_verbs.head()
```

```python
def regressive_e_raising_backing(word):
    pattern1 = '\*ne\*\-[^ieɛaøuoʌ]+u'
    pattern2 = 'me\-[^ieɛaøuoʌ]+u'
    pattern3 = '(?:be|de)+[^ieɛaøuoʌ]+u'

    if match:=re.search(pattern1, word):
        index = match.span()[0]
        word = word[:index+2] + 'u' + word[index+3:]  
    if match:=re.search(pattern2, word): #use if insted of elif because I want this also to apply to the verbs that were checked above
        index = match.span()[0]
        word = word[:index+1] + 'u' + word[index+2:]
    if match:=re.search(pattern3, word):
        index = match.span()[0]
        word = word[:index+1] + 'u' + word[index+2:]
    return word
print(regressive_e_raising_backing('me-*ne*-gggguz-em'))
print(regressive_e_raising_backing('*ne*-guz-em'))
print(regressive_e_raising_backing('me-guz-*em'))
print(regressive_e_raising_backing('*be*-guz-*em'))
all_verbs['Generated Verb-SR'] = all_verbs['Generated Verb-SR'].apply(regressive_e_raising_backing)
all_verbs.to_csv('./all_verbs_reg_e_raising_backing.csv', encoding = 'utf-8-sig', index = False)
all_verbs = pd.read_csv('all_verbs_reg_e_raising_backing.csv')
all_verbs.head()
```

```python
def a_raising(word):
    pattern1 =  'a[^iaøuoʌe]*i' # /e/ cannot block this harmony me-lɛɢ-den-ɛst-i
    pattern2 = 'u\-mɛ\-'
    pattern3 = '^\*mɛ\*\-'
    pattern4 = '\*ma\*\-[iaeøuoʌ]'
    while match:=re.search(pattern1, word): 
        index = match.span()[0]
        word = word[:index] + 'ɛ' + word[index+1:]
    if match:=re.search(pattern2, word):    
        word = re.sub(pattern2,'u-ma-',word) # exceptional case for /a/ raising
    if match:=re.search(pattern3, word):    
        word = re.sub(pattern3,'*ma*-',word) 
    if match:=re.search(pattern4, word):
        index = match.span()[0]
        word = word[:index+5] + '' + word[index+6:]         
    return word
print(a_raising('ma-ne-ra͡s-ikjhk'))
print(a_raising('a-*ne*-ger-i-ast-aiʃ(ɛ)'))
print(a_raising('me-*ne*-and͡ʒan-i-em(ɛ)'))
print(a_raising('m-and͡ʒan-i-*jast'))
print(a_raising('be-baχʃ-ast-*avim(ɛ)*'))
print(a_raising('m-*n*-ajsen-i-jem(ɛ)'))
print(a_raising('ma-*n*-ajsen-i-jem(ɛ)'))
print(a_raising('*ma*-iduʃ'))
print(a_raising('a-ger-i-jast-*abe*'))


all_verbs['Generated Verb-SR'] = all_verbs['Generated Verb-SR'].apply(a_raising)
all_verbs.to_csv('./all_verbs_a_raising.csv', encoding = 'utf-8-sig', index = False)
all_verbs = pd.read_csv('all_verbs_a_raising.csv')
all_verbs.head()
```

```python

def stress_changer(word):
    match = re.search('\*.+\*', word)
    star_start = match.span()[0]
    star_end = match.span()[1]
    starred_portion = word[star_start:star_end]
    
    word = word[:star_start] + star_changer(starred_portion) + word[star_end:]
    return word


def star_changer(starred_portion):
    starred_portion = starred_portion.replace('*', '') # first take starts off a string
    if starred_portion in starred_dict:
        return(starred_dict[starred_portion]) # after return, functions stopps running
    length = len(vowels.findall(starred_portion))
    if length == 0: 
        starred_portion = '*' + starred_portion + '*'
        return starred_portion
        return starred_portion
    if length == 1: # if there is only one vowel in the string
        start,end = vowels.search(starred_portion).span() #unpacks the tuple
        starred_portion = starred_portion[:start]  + '*' + starred_portion[start] + '*' + starred_portion[end:]
        return starred_portion
    if length !=1: # can be 2 or 3
        starred_portion =  starred_portion[::-1] # we are interested in the last vowel. An indirect way to do this 
        # is to reverse the order of the starred portion, first the first vowel(which will be the last one we are interested it, 
        # and then reverse it back again to take it back to its original order)
        start,end = vowels.search(starred_portion).span() #unpacks the tuple
        starred_portion = starred_portion[:start]  + '*' + starred_portion[start] + '*' + starred_portion[end:]
        starred_portion =  starred_portion[::-1] # reverse back
        return starred_portion
    


all_verbs['Generated Verb-SR'] = all_verbs['Generated Verb-SR'].apply(stress_changer)
all_verbs.to_csv('./all_verbs_stress_changer.csv', encoding = 'utf-8-sig', index = False)
all_verbs = pd.read_csv('all_verbs_stress_changer.csv')
print(all_verbs.head(5))     

        
print(stress_changer('*ne*'))
print(stress_changer('*ma*'))
print(stress_changer('*em(ɛ)*'))
print(stress_changer('*eʃ(ɛ)*'))
print(stress_changer('*ɛjsend*'))
print(stress_changer('*oχʌrd*'))
print(stress_changer('*idut*')) 
print(stress_changer('*endɛ*'))
print(stress_changer('*ijʌ*')) 
print(stress_changer('bekɛt*ijʌ*'))
print(stress_changer('*n*-a'))

```

```python
def indʒ_fix(word):
    pattern1 = 'min\*i\*nd͡ʒ'
    pattern2 = 'mind͡ʒ'
    pattern3 = 'b[\*]?i[\*]?nd͡ʒ'
    pattern4 = 'n\*i\*nd͡ʒ'
    if match:=re.search(pattern1, word):
        word = re.sub(pattern1,'men*e*nd͡ʒ',word)
    if match:=re.search(pattern2, word):
        word = re.sub(pattern2,'mend͡ʒ',word)
    if match:=re.search(pattern3, word):
        word = re.sub(pattern3,'b*e*nd͡ʒ',word)
    if match:=re.search(pattern4, word):
        word = re.sub(pattern4,'n*e*nd͡ʒ',word)
    return(word) 
all_verbs['Generated Verb-SR'] = all_verbs['Generated Verb-SR'].apply(indʒ_fix)
all_verbs.to_csv('./all_verbs_indʒ_fix.csv', encoding = 'utf-8-sig', index = False)
all_verbs = pd.read_csv('all_verbs_indʒ_fix.csv')
all_verbs.head()

print(indʒ_fix('mind͡ʒ'))
print(indʒ_fix('bind͡ʒ'))
print(indʒ_fix('n*i*nd͡ʒ'))
print(indʒ_fix('min*i*nd͡ʒ'))
```

```python
def e_raising_j(word):
    pattern1 = 'ej'
    pattern2 = 'e\*j'
    if match:=re.search(pattern1, word):
        word = re.sub(pattern1, 'ij', word)
    if match:=re.search(pattern2, word):
        word = re.sub(pattern2, 'i*j', word)
    return word
print(e_raising_j('ʌn*e*jɛrijɛstɛbimon'))

all_verbs['Generated Verb-SR'] = all_verbs['Generated Verb-SR'].apply(e_raising_j)
all_verbs.to_csv('./all_verbs_e_raising_j.csv', encoding = 'utf-8-sig', index = False)
all_verbs = pd.read_csv('all_verbs_e_raising_j.csv')
```
